```python
import pandas as pd

# 1. 载入并合并 17 个文件
files = [f"synthetic_eval_{i}.csv" for i in range(1, 18)]
dfs = [pd.read_csv(f) for f in files]
df = pd.concat(dfs, ignore_index=True)

# 2. 删除 GPT-3.5 相关的两列
df = df.drop(columns=["GPT-3.5_Score", "GPT-3.5_Explain"], errors="ignore")

# 3. 输出为 eval_sample.csv
df.to_csv("eval_sample.csv", index=False)

print("合并后的数据已保存为 eval_sample.csv,形状:", df.shape)
```
